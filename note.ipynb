```python
import matplotlib.pyplot as plt
import numpy as np

# Data for plotting
t = np.arange(0.0, 20.0, 0.01)

s = np.exp(-t) * np.sin(t) * np.exp(0.5 * t) 

fig, ax = plt.subplots()
ax.plot(t, s)

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='')
ax.grid()

# fig.savefig("test.png")
plt.show()
```
